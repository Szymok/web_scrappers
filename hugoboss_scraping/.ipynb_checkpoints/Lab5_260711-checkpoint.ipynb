```python
# 2. Przygotuj skrypt w języku Python do pobierania informacji o pracownikach Wydziału Zarządzania.
import requests

r=requests.get("https://wzr.ug.edu.pl/wydzial/index.php?str=121")
from bs4 import BeautifulSoup
tagi=BeautifulSoup(r.text, 'html.parser')
nazwiska=tagi.findAll('div', attrs={'class':'pracownik_b'})
for nazwisko in nazwiska:
    print(nazwisko.text.strip())
```

```python
# 3. Zapisz wyniki w:
# - pliku CSV,
# - bazie danych,
# - JSON.

plikCSV=open("plik260711.csv","w")
plikJSON=open("plik260711.json","w")
plikJSON.write('[')
for nazwisko in nazwiska:
    plikCSV.write(nazwisko.text.strip()+"\n")
    plikJSON.write("{'nazwisko':"+nazwisko.text.strip()+"'},")
    kursor.execute("insert into prac_260711 values ('"
                   +nazwisko.text.strip()[:40]+"')")
    polaczenie.commit()
plikJSON.write(']')    
print("zapisano dane w csv, json, sql")
```

```python
plikJSON.close()
plikCSV.close()
```
